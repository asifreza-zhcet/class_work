```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\rezaa\Videos\Captures\11.Machine Learning\2.Classification\14_march\car_sales_classification_dartaset.csv")
dataset.head()
```

```python
def model(ts=.2,rs=0,pen='l2',sol='lbfgs'):

        
    
    X = dataset.iloc[:,2:4].values
    y = dataset.iloc[:,-1].values
    
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=rs, test_size=ts)
    
    
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    from sklearn.linear_model import LogisticRegression
    classifier = LogisticRegression(penalty=pen,solver=sol)
    classifier.fit(X_train,y_train)
    y_pred = classifier.predict(X_test)
    
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_test, y_pred)
    cm
    
    from sklearn.metrics import accuracy_score
    ac = accuracy_score(y_test, y_pred)
    ac
    
    from sklearn.metrics import classification_report
    cr = classification_report(y_test, y_pred)
    cr
    
    bias = classifier.score(X_train, y_train)
    bias
    
    variance = classifier.score(X_test, y_test)
    variance
    
    result = [ts, rs, pen, sol, ac, bias, variance ]
    
    
    return result
```

```python
tab = []
tab.append(model(ts=.25,rs=0))
tab.append(model(ts=.25,rs=41))
tab.append(model(ts=.25,rs=51))
tab.append(model(ts=.20,rs=0))
tab.append(model(ts=.20,rs=41))
tab.append(model(ts=.20,rs=51))
tab.append(model(ts=.25,pen='l1',sol='liblinear'))
tab.append(model(ts=.25,pen='l1',sol='saga'))

tab.append(model(ts=.25,pen='l2',sol='lbfgs'))
tab.append(model(ts=.25,pen='l2',sol='liblinear'))
tab.append(model(ts=.25,pen='l2',sol='newton-cg'))
tab.append(model(ts=.25,pen='l2',sol='newton-cholesky'))
tab.append(model(ts=.25,pen='l2',sol='sag'))
tab.append(model(ts=.25,pen='l2',sol='saga'))
```

```python
df = pd.DataFrame(tab, columns =['test size','random state','penalty','solver','accuracy','bias','variance'])
df
```
