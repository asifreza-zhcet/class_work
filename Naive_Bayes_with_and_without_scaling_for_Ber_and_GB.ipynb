# Naive Bayes classification

## Importing libraries

```python
import numpy as np
import pandas as pd
```

## Importing datset

```python
dataset = pd.read_csv(r'C:\Users\rezaa\Videos\Captures\11.Machine Learning\2.Classification\20th- Naive bayes\Social_Network_Ads.csv')
X = dataset.iloc[:,2:4].values
y = dataset.iloc[:,-1].values                    
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=0)
```

## 1. Training data with Bernauli NB algo

### Without standard scaler

```python
from sklearn.naive_bayes import BernoulliNB
model_Ber = BernoulliNB()
model_Ber.fit(X_train,y_train);
```

### Plotting the decision region

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train, y_train, clf=model_Ber)
```

### Accuracy and confusion matrix

```python
y_pred = model_Ber.predict(X_train)
from sklearn.metrics import confusion_matrix, accuracy_score
print(f'the confusion matrix\n{confusion_matrix(y_train, y_pred)}')
print(f'the accuracy score {accuracy_score(y_train, y_pred)}')
```

### Conclusion:
The results show the model is heavily underfited

## With standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
model_Ber_sc = BernoulliNB()
model_Ber_sc.fit(X_train_sc,y_train)
plot_decision_regions(X_train_sc, y_train, clf=model_Ber_sc)
```

### Accuracy and confusion matrix

```python
X_test_sc = sc.transform(X_test)
y_pred_Ber_sc = model_Ber_sc.predict(X_test_sc)
print(f'the confusion matrix\n{confusion_matrix(y_test, y_pred_Ber_sc)}')
print(f'the accuracy score {accuracy_score(y_test, y_pred_Ber_sc)}')
```

### Conclusion
Scaling not only increased the accuracy but also made the model non underfit

## 2. Trainning data with Gaussian NB algo

### Without scaling

```python
from sklearn.naive_bayes import GaussianNB
model_GB = GaussianNB()
model_GB.fit(X_train,y_train);
```

### Ploting the decision region

```python
plot_decision_regions(X_train, y_train, clf=model_GB);
```

### Accuracy and confusion matrix

```python
y_pred_GB = model_GB.predict(X_test)
print(f'the confusion matrix\n{confusion_matrix(y_test, y_pred_GB)}')
print(f'the accuracy score {accuracy_score(y_test, y_pred_GB)}')
```

### With standard scaler

```python
model_GB_sc = GaussianNB()
model_GB_sc.fit(X_train_sc,y_train)
plot_decision_regions(X_train_sc, y_train, clf=model_GB_sc);
```

### Accuracy and confusion matrix

```python
y_pred_GB_sc = model_GB_sc.predict(X_test_sc)
print(f'the confusion matrix\n{confusion_matrix(y_test, y_pred_GB_sc)}')
print(f'the accuracy score {accuracy_score(y_test, y_pred_GB_sc)}')
```
